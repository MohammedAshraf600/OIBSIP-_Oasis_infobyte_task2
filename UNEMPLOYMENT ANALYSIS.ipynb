# data preprocessing

```python
#check duplicated values
df.duplicated().sum()
```

```python
#let's drop duplicated values
df=df.drop_duplicates()
```

```python
#check after drop duplicated values
df.duplicated().sum()
```

```python
#checking our data if there are nulls or not 
df.isnull().sum()
```

```python
df
```

```python
# we will deal with miss values by Mean&Mode&Median
# Filling missing values with the mean of each respective column
df[' Estimated Unemployment Rate (%)'] = df[' Estimated Unemployment Rate (%)'].fillna(df[' Estimated Unemployment Rate (%)'].mean())
df[' Estimated Employed'] = df[' Estimated Employed'].fillna(df[' Estimated Employed'].mean())
df[' Estimated Labour Participation Rate (%)'] = df[' Estimated Labour Participation Rate (%)'].fillna(df[' Estimated Labour Participation Rate (%)'].mean())
```

```python
df['Region'] = df['Region'].fillna(df['Region'].mode().iloc[0])
df[' Date'] = df[' Date'].fillna(df[' Date'].mode().iloc[0])
df['Area'] = df['Area'].fillna(df['Area'].mode().iloc[0])
df[' Frequency'] = df[' Frequency'].fillna(df[' Frequency'].mode().iloc[0])
```

```python
#checking our data after dealing with null_values
df.isnull().sum()
```

```python
#split Date into Day&month&year 
df[' Date'] = pd.to_datetime(df[' Date'])
# Extracting day, month, and year into new columns using .loc
df['day'] = df[' Date'].dt.day
df['month'] = df[' Date'].dt.month
df['year'] = df[' Date'].dt.year
```

```python
#my data after spiltting Date
df.head()
```

```python
#change the type of Estimated Unemployment Rate & Estimated Labour Participation Rate & Estimated Employed from float to int
df[' Estimated Unemployment Rate (%)'] = df[' Estimated Unemployment Rate (%)'].astype(int)
df[' Estimated Labour Participation Rate (%)'] = df[' Estimated Labour Participation Rate (%)'].astype(int)
df[' Estimated Employed'] = df[' Estimated Employed'].astype(int)
```

# let's detect outliers

```python
#visualizatoin to see the outliers

fig, ax = plot.subplots(2, 2, figsize=(8, 12))

# Boxplot for Estimated Unemployment Rate (%)
ax[0, 0].boxplot(x=df[[' Estimated Unemployment Rate (%)']])
ax[0, 0].set_title('Estimated Unemployment Rate (%)')

# Boxplot for Estimated Employed
ax[0, 1].boxplot(x=df[[' Estimated Employed']])
ax[0, 1].set_title('Estimated Employed')

# Boxplot for Estimated Labour Participation Rate (%)
ax[1, 0].boxplot(x=df[[' Estimated Labour Participation Rate (%)']])
ax[1, 0].set_title('Estimated Labour Participation Rate (%)')

# Remove the last subplot (ax[1, 1]) from the layout
fig.delaxes(ax[1, 1])

plot.show()
```

```python
#Q1 and Q3 calculations, and the interquartile range (IQR) for the 'Estimated Unemployment Rate (%)'
Q1_Estimated_Unemployment_Rate=df[' Estimated Unemployment Rate (%)'].quantile(0.25)
Q3_Estimated_Unemployment_Rate=df[' Estimated Unemployment Rate (%)'].quantile(0.75)
IQR_Estimated_Unemployment_Rate=Q3_Estimated_Unemployment_Rate - Q1_Estimated_Unemployment_Rate
print(IQR_Estimated_Unemployment_Rate)
```

```python
df = df [~((df[' Estimated Unemployment Rate (%)']<(Q1_Estimated_Unemployment_Rate-1.5*IQR_Estimated_Unemployment_Rate))|(df[' Estimated Unemployment Rate (%)']>(Q3_Estimated_Unemployment_Rate+1.5*IQR_Estimated_Unemployment_Rate)))]
```

```python
#let's visualize after detecting outliers in Estimated Unemployment Rate (%)
sns.boxplot(x=' Estimated Unemployment Rate (%)', data=df)
```

```python
#Q1 and Q3 calculations, and the interquartile range (IQR) for the 'Estimated Employed'
Q1_Estimated_Employed=df[' Estimated Employed'].quantile(0.25)
Q3_Estimated_Employed=df[' Estimated Employed'].quantile(0.75)
IQR_Estimated_Employed=Q3_Estimated_Employed - Q1_Estimated_Employed
print(IQR_Estimated_Employed)
```

```python
df = df [~((df[' Estimated Employed']<(Q1_Estimated_Employed-1.5*IQR_Estimated_Employed))|(df[' Estimated Employed']>(Q3_Estimated_Employed+1.5*IQR_Estimated_Employed)))]
```

```python
#let's visualize after detecting outliers in Estimated Employed
sns.boxplot(x=' Estimated Employed', data=df)
```

```python
#Q1 and Q3 calculations, and the interquartile range (IQR) for the 'Estimated Labour Participation Rate (%)'
Q1_Estimated_Labour_Participation_Rate=df[' Estimated Labour Participation Rate (%)'].quantile(0.25)
Q3_Estimated_Labour_Participation_Rate=df[' Estimated Labour Participation Rate (%)'].quantile(0.75)
IQR_Estimated_Labour_Participation_Rate=Q3_Estimated_Labour_Participation_Rate - Q1_Estimated_Labour_Participation_Rate
print(IQR_Estimated_Labour_Participation_Rate)
```

```python
df = df [~((df[' Estimated Labour Participation Rate (%)']<(Q1_Estimated_Labour_Participation_Rate-1.5*IQR_Estimated_Labour_Participation_Rate))|(df[' Estimated Labour Participation Rate (%)']>(Q3_Estimated_Labour_Participation_Rate+1.5*IQR_Estimated_Labour_Participation_Rate)))]
```

```python
#let's visualize after detecting outliers in Estimated Labour Participation Rate (%)
sns.boxplot(x=' Estimated Labour Participation Rate (%)', data=df)
```
